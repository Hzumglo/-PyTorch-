# 1.线性代数
## 1.1 标量
标量由只有一个元素的张量表示。

```python
import torch
x = torch.tensor(3.0)
y = torch.tensor(2.0)

x + y, x * y, x / y, x ** y
```

## 1.2 向量
向量可以被视为标量值组成的列表。这些标量值被称为向量的**元素或分量**。在数学上，具有一个轴的张量表示向量。一般来说张量可以具有任意长度。

```python
x = torch.tensor([0, 1, 2, 3])
x
```

```python
# 可以使用下标来引用向量的任一元素
x[3]
```

向量只是一个数字数组，就像每个数组都有长度一样，向量也是如此。向量的长度通常称为**向量的维度**。

```python
# 与普通数组一样，我们可以用len来访问张量的长度
# 当张量只用一个轴表示一个向量时，我们也可以通过.shape属性来访问向量的长度
x.shape, len(x)
```

**向量或轴的维度被用来表示向量或轴的长度，而张量的维度用来表示张量具有的轴数。所以张量某个轴的维度就是这个轴的长度。**
## 1.3 矩阵
向量将标量从零阶推广到一阶，矩阵可以将向量从一维推广到二维，在代码中表示为具有两个轴的张量。</br>
当矩阵具有相同的列数和行数称之为**方阵**。

```python
A = torch.arange(20).reshape(5, 4)
A
```

可以通过行索引和列索引来访问矩阵中的标量元素$a_{ij}$

```python
A[1][1]
```

当我们交换矩阵的行列时，结果称为矩阵的**转置**。用$A^T$表示。

```python
A.T
```

作为方阵的一种特殊类型，**对称矩阵**A等于其转置：$A=A^T$

```python
B = torch.tensor([[1, 2 , 3], [2, 0, 4], [3, 4, 5]])
B, B == B.T
```

## 1.4 张量
**张量**是指描述具有任意数轴的n维数组的通用方法。（向量时一阶张量，矩阵时二阶张量）

```python
X = torch.arange(24).reshape(2, 3, 4)
X
```

## 1.5 张量算法的基本性质

```python
# 给定同样形状大小的张量，任何按元素二元运算的结果都将是相同形状的张量
A = torch.arange(20, dtype = torch.float32).reshape(5, 4)
B = A.clone()
A, A + B
```

两个矩阵的按元素乘法称为**哈达玛积($\odot $)，即矩阵乘法**，代码如下所示：

```python
A * B
```

将张量加上或乘以一个**标量**不会改变张量的形状，其中张量的每个元素都将与标量相加或相乘。

```python
a = 2
X = torch.arange(24).reshape(2, 3, 4)
a + X, (a * X).shape
```

## 1.6 降维
我们可以对任意张量进行的一个有用操作就是计算其元素之和。

```python
x = torch.arange(4, dtype = torch.float32)
x, x.sum()
```

```python
A.shape, A.sum()
```

默认情况下，**调用求和函数会沿所有的轴降低张量的维度，使它变为一个标量。** 我们还可以指定张量沿哪一个轴来通过求和降低维度。

```python
A_sum_axis0 = A.sum(axis = 0)
A_sum_axis0, A_sum_axis0.shape, A.sum(axis = [0, 1])
```

```python
# 求均值
A.mean(), A.sum() / A.numel()
```

```python
# 非降维求和
sum_A = A.sum(axis = 1, keepdims = True)
A, sum_A
```

```python
A / sum_A
# 会采用广播机制，对sum_A进行扩展，用的复制法
```

```python
# .cumsum() 可以求某个轴元素的累积求和
A, A.cumsum(axis = 0)
```

## 1.7 点积
给定两个向量$x,y \in R^d$，他们的点积$x^Ty或(<x,y>)$。是相同位置的按元素乘积的和：$x^Ty = \sum_{i=1}^d x_iy_i$。

```python
y = torch.ones(4, dtype = torch.float32)
x, y, torch.dot(x, y)
```

```python
torch.sum(x * y)
```

## 1.8 矩阵-向量积
我们之前定义的矩阵$A \in R^{m \times n}$和向量$x \in R^n$。我们可以将矩阵A用其行向量表示。矩阵向量积$Ax$是一个长度为m的列向量。</br>
**一个矩阵乘以一个列向量相当于矩阵的列向量的线性组合；** </br>
**一个行向量乘以矩阵，相当于矩阵的行向量的线性组合。** </br>
代码中可以用**mv函数可以求矩阵-向量积**

```python
A.shape, x.shape, torch.mv(A, x)
```

## 1.9 矩阵-矩阵乘法 
$A \in R^{x \times y}, B \in R^{y \times z}$，能乘的条件就是，前面矩阵的列数与后面矩阵的行数得一致。代码中可以用**mm函数求矩阵-矩阵乘法**。</br>
**矩阵-矩阵乘法可以简单地称为矩阵乘法，不要与哈达玛积混淆。**

```python
B = torch.ones(4, 3)
A.shape, B.shape, torch.mm(A, B)
```

## 1.10 范数
线性代数中最有用的一些运算符是**范数**。非正式地说，向量的范数表示一个向量有多大，这里说的是大小。</br>
在线性代数中，向量范数是将**向量映射到标量的函数$f$**，向量范数具有一些性质：</br>
$\cdot$ 如果我们按常数因子$\alpha$缩放向量的所有元素，其范数也会按相同常数因子的绝对值缩放：$f(\alpha x) = |\alpha|f(x)$</br>
$\cdot$ 三角不等式：$f(x + y) \le f(x) + f(y)$ </br>
$\cdot$ 范数必须是非负的：$f(x) \ge 0$ </br>

**范数很像距离的度量。**</br>
假设n维向量x中的元素是$x_1, x_2,...,x_n$，其 **$L_2$范数是向量元素平方和的平方根**：
$$
||x||_2=\sqrt{\sum_{i=1}^n x_i^2}
$$
$L_2$范数中常常省略下标2。代码如下：

```python
u = torch.tensor([3.0, -4.0])
torch.norm(u)
# 可以理解为二维平面给上的点到原点的距离
```

除了$L_2$范数，也会经常遇到$L_1$范数，**它表示为向量的绝对值之和：**
$$
||x||_1=\sum_{i=1}^n|x_i|
$$
与$L_2$范数相比，$L_1$范数受异常值的影响较小。

```python
torch.abs(u).sum()
```

$L_2$和$L_1$范数都是更一般的$L_p$范数的特例：
$$
||x||_p = (\sum_{i = 1}^n|x_i|^p)^{\frac{1}{p}}
$$

类似于向量的$L_2$范数，矩阵$X \in R^{m \times n}$的**佛罗贝尼乌斯范数**是矩阵元素平方和的平方根：
$$
||X||_F=\sqrt{\sum_{i=1}^M \sum_{j=1}^n x_{ij}^2}
$$
佛罗贝尼乌斯范数具有向量范数的所有性质，他就是像是矩阵向量的$L_2$范数，

```python
torch.norm(torch.ones(4, 9))
```

### 范数和目标
在深度学习中，我们经常试图解决优化问题：**最大化分配给观测数据的概率；最小化预测值和真实观测值之间的距离。** 用向量表示物品，以便最小化相似项目之间的距离，最大化不同项目之间的距离。**除了数据目标或许是深度学习算法最重要的组成部分，通常被表达为范数。（也就是我们通常定义的损失函数也就是定义成范数的形式）。**
